# Data cleaning

The goal of data cleaning is to prepare the student performance dataset for analysis and modeling.
This step includes loading the data, checking for missing values, correcting data types and very basic exploration.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/StudentPerformanceFactors.csv")
df
```

```python
df.info()
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
missing_percentage = df.isnull().sum() / len(df) * 100
missing_percentage
```

```python
for col in ["Teacher_Quality", "Distance_from_Home", "Parental_Education_Level"]:
    mode_value = df[col].mode()
    df[col] = df[col].fillna(mode_value[0] if not mode_value.empty else 'Unknown')
```

```python
df.isnull().sum()
```

The amount of missing values in the dataset is very low, all below 1.5% of all rows. Given the small percentage of the missing data, a simple, transparent imputantion was chosen while still preserving the structure of the dataset.

1. Teacher Quality
It is an ordinal variable representing teacher effectivness. The chosen imputation method is a mode, most frequent value. Since the missing rate is very low, filling with the most common value maintains the natural distribution of the dataset.

2. Parental Education Level
It is a categorical variable representing the highest education level of parents. Once again, the mode was used as an imputation method, beacuse of the very low missing rate.

3. Distance from Home
It is treated as ordinal variable represting relative distance from home to school. Again, the method of imputation was a mode. Since the missing values rate was so low and the variable is treated as ordinal rather than numeric, the mode is the most fitting choice.
